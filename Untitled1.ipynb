```python
tmp = ""
def getDate(h):
    global tmp
    try :
        tmp = h.parent.find_previous_sibling('td').get_text()
        return tmp
    except:
        return ""


def weekday_(w):
    prpr = {      
        '월': '0',
        '화': '1',
        '수': '2',
        '목': '3',
        '금': '4',
        '토': '5',
        '일': '6',
    }
    for key, value in prpr.items():
        w = w.replace(key, value)
    return w    
    
    
    
    
def month_(m):
    prpr = {     
        '4': '04',
        '5': '05',
        '6': '06',
        '7': '07',
        '8': '08',
        '9': '09',
    }
    for key, value in prpr.items():
        m = m.replace(key, value)
    return m    
    
    
    
def LOCATION_(s):
    prpr = {      
        '잠실': '서울',
        '사직': '부산',
        '고척': '서울',
        '문학': '인천',
        '마산': '창원',
    }
    for key, value in prpr.items():
        s = s.replace(key, value)
    return s    



def team(s):
    prpr = {      
        'kt': 'KT',
        '넥센': '키움',
    }
    for key, value in prpr.items():
        s = s.replace(key, value)
    return s 
```
